<h3>각 실험 결과 평균 계산</h3>
<hr>

```python
synchronization_cnt_tot = []
```

```python
for i in range(100, 14100, 100):
    temp = synchronization[(synchronization['Num of block'] == i)]
    temp = temp[(temp['Count'] != 1)]
    
    synchronization_cnt_tot.append(temp.mean().Time)
```

<h3>실험 결과 그래프</h3>
<hr>

```python
plt.plot(synchronization_cnt_10['Num of block'], synchronization_cnt_tot)

plt.xlabel('Number of block')
plt.ylabel('Synchronization Time (ms)')

plt.show()
```
